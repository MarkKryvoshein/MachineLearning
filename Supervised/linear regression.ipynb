```python
import numpy
import pandas as pd
```

```python
data = {
    'Days':[1,2,3,4,5],
    'Coins':[2,5,6,8,15]
}

df = pd.DataFrame(data)
df.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(x=df.Days, y=df.Coins, color='blue')

plt.title("Графік залежності")
plt.xlabel("Дні")
plt.ylabel("Коіни")

plt.show()
```

```python
from sklearn.linear_model import LinearRegression

days = list(map(int, input().split()))
coins = list(map(float, input().split()))

print(days)


data_= {
    'Days': days,
    'Coins': coins

}

data_frame = pd.DataFrame(data_)

X = data_frame[['Days']]
Y = data_frame['Coins']

model = LinearRegression()

model.fit(X, Y)

print("Вага[W]", model.coef_)
print("Зміщення", model.intercept_)

Y_pred = model.predict(X)

plt.scatter(X, Y, color="blue")
plt.plot(X, Y_pred, color="red")

plt.show()
```

```python
new_value = [[10]]
predicted_coins = model.predict(new_value)

print(f'Прогноз для {new_value[0][0]} дня: {predicted_coins[0]}')
```
