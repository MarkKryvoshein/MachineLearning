```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies_df.dropna(inplace=True)
movies_df.info()
```

```python
movies_df['homepage'].fillna('https', inplace=True)
movies_df['homepage'].head()
```

```python
movies_df['release_year'] = pd.to_datetime(movies_df['release_date']).dt.year
```

```python
genre_counts = movies_df['genres'].value_counts()
print(genre_counts)
```

```python
import json
from collections import Counter

def parse_genre(element):
  try:
    return [parsed['name'] for parsed in json.loads(element.replace("'", '"'))]
  except:
    return []
parsed_frame = [el for halflist in movies_df['genres'].apply(parse_genre) for el in halflist ]
genre_counts = Counter(parsed_frame)
genre_counts = dict(genre_counts.most_common())
#genre_counts = dict(genre_counts.items().sort())
sns.barplot(x=list(genre_counts.keys()), y =list(genre_counts.values()))

plt.title('Кількість фільмів за жанрами')
plt.xlabel('Жанр')
plt.ylabel('Кількість фільмів')

plt.xticks(rotation=90)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast
```

```python
def extract_genres(genres_str):
  try:
    genres = ast.literal_eval(genres_str)
    return [genre['name'] for genre in genres]
  except ValueError:
    return []
```

```python
try:
  movies_df['genres'] = movies_df['genres'].apply(extract_genres)
except:
  movies_df['genres'] = []
```

```python
genre_counts = movies_df['genres'].value_counts()
print(genre_counts)
```

```python
genres_exploded = movies_df[['title', 'release_date', 'budget', 'revenue', 'genres']].explode('genres')
# print(genres_exploded)

genre_count = genres_exploded['genres'].value_counts()
# print(genre_count)

# Візуазізація жанрів
plt.figure(figsize=(10,6))
sns.barplot(x=genre_count.index, y=genre_count.values)
plt.title("Кількість фільмів за жанрами")
plt.xlabel("Жанр")
plt.ylabel("Кількість фільмів")
plt.xticks(rotation=45)
plt.tight_layout()
# plt.show()
```

```python
movies_df['budget'] = pd.to_numeric(movies_df['budget'], errors='coerce')

movies_df['revenue'] = pd.to_numeric(movies_df['revenue'], errors='coerce')
```

```python
movies_df.fillna({'budget': 0, 'revenue': 0}, inplace=True)
```

```python
movies_df['profit'] = movies_df['revenue'] - movies_df['budget']
```

```python
print(movies_df['profit'])
```

```python

sns.scatterplot(x=movies_df['budget'], y =movies_df['profit'])

plt.title('Профіт vs бюджет')
plt.xlabel('Бюджет')
plt.ylabel('Профіт')

plt.xticks(rotation=90)

plt.show()
```
